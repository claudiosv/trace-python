```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv = pd.read_csv('notebooks/csv_projs.csv')
```

```python
csv.plot()
```

```python
csv
```

```python
csv.sort_values(by='test cases', ascending=False)
```

```python
csv['ratio'] = csv.apply(lambda row: (row['test cases']/(max(1,row['test suites']))), axis=1)
```

```python
(csv.sort_values(by='test cases', ascending=False)
   .reset_index(drop=True)[:25]
   .plot.barh(x='project name', y='test cases', figsize=(20, 20)))

plt.ylabel('Project')
plt.xlabel('Test cases')

plt.show()

```

```python
(csv.sort_values(by='ratio', ascending=False)
   .reset_index(drop=True)[:25]
   .plot.barh(x='project name', y='ratio', figsize=(20, 20)))
plt.ylabel('Project')
plt.xlabel('Mean test case count per test suite')

plt.show()


```

```python
(csv.sort_values(by='test suites', ascending=False)
   .reset_index(drop=True)[:25]
   .plot.barh(x='project name', y='test suites', figsize=(20, 20)))


plt.ylabel('Project')
plt.xlabel('Test suites')

plt.show()
```
